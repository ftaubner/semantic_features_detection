```python
SAVE_DIR = '/home/erbachj/MaskRCNN/Mask_RCNN/data/InteriorNet/data/HD7'
subset = 'train'
names = {'train':[], 'val':[], 'test':[]}

counter = 0
limit = 100
with open(os.path.join('/home/erbachj/MaskRCNN/Mask_RCNN/data/InteriorNet', 'listing_of_folder_HD7.csv'), newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        name = row[0]
        link = row[1]
        gdown.download(link,os.path.join(SAVE_DIR, subset, name))
        names[subset].append(name)
        counter += 1
        if counter >= 160:
            break
        elif counter >= 130:
            subset = 'test'
        elif counter >= 100:
            subset = 'val'
pd.DataFrame(names).to_csv(SAVE_DIR, index = False)
```
